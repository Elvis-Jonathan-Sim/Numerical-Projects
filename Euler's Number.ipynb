```python
import numpy as np
import matplotlib.pyplot as plt

#Making Euler Sequence  
E = []
n = 100
for i in range(1,(n+1)):
    e = (1 + (1/i))**i
    E.append(e)
    
x = np.arange(1,(n+1))
plt.figure(figsize=(20,10))
plt.scatter(x,E)
plt.grid()
```

```python
plt.figure(dpi=150)
plt.bar(x,E)
```

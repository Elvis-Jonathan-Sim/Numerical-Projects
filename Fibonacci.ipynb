## Dynamic Programming

### Fibonacci Sequence

```python
import numpy as np
import matplotlib.pyplot as plt

def Fibo(n):
    Fibo = [0,1,1]
    x1 = 1
    x2 = 1
    for i in range(0, n-2):
        x3 = x1 + x2
        x1 = x2
        x2 = x3
        Fibo.append(x3)
        
    return Fibo
```

```python
Fibo(10)
```

```python
x = np.arange(0, len(Fibo(10)))
plt.figure(figsize=(15,10))
plt.scatter(x,Fibo(10))
plt.grid()
plt.xlabel("Iterations")
plt.ylabel("Fibonacci Value")
plt.show()
```

## Recursive Function 

<br> Creating a straight forward fibonnaci Sequence

```python
def fib(n):
    if n == 0:
        result = 0
    if n==1 or n==2:
        result = 1
    else:
        result = fib(n-1) + fib(n-2)
    return result
```

```python
fib(10)
```

```python
fib(1000)
```

<br> Calculating the Fibonacci for ex: 1000 Iterations requires a long time to be computed. 
<br> Here we see that the **Time Complexity** = *O(2^n)*

## Memoization

<br> **Memoization** is a type of dynamic programming which divide the problem into *sub-problems*.
<br> Memoization reduces the **Time Complexity** to << *O(2^n)*. 

```python
def fib_2(n, memo):
    if memo[n] is not None: 
        return memo[n]
    if n==1 or n==2:
        return 1
    else:
        result = fib_2(n-1,memo) + fib_2(n-2,memo)
    memo[n] = result
    return result

def fib_memo(n):
    memo = [None] * (n+1)
    return fib_2(n, memo)
```

```python
fib_memo(10)
```

```python
fib_memo(100)
```

## Bottom Up 

<br> Bottom Up offers a better solution to the Fibonacci recursive problem. 
<br> The **Time Complexity** is reduced from *O(2^n)* to approximately *O(n)*.

```python
def fib_bottom_up(n):
    if n == 1 or n==2:
        return 1
    bottom_up = [None] * (n+1)
    bottom_up[1] = 1
    bottom_up[2] = 1
    for i in range(3, n+1):
        bottom_up[i] = bottom_up[i-1] + bottom_up[i-2]
    return bottom_up[n]
```

```python
fib_bottom_up(10)
```

```python
fib_bottom_up(1000)
```

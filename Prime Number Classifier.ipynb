## Prime Number Classifier

<br> The Idea of Prime Numbers: Numbers that are divisible by itself and one. 
<br> *Implementation:* Using Modulo Arithmetic to classify if a number is a prime number. If the mod(n) = 0 for a certain value (n) of iterations, then the number is not a prime number

```python
def isPrime(number):
    
    if type(number) != type(1):
        raise ValueError("Please Give an Integer")
    """
    if number == 1:
        print(" It is an Identity Number! ")
    elif number == 2:
        print("2 is a Prime Number")
    elif number <= 0:
        raise ValueError("Please Give an Integer larger than 0") """
    for i in range(2, number):
        if number % i == 0:
            return False
            print(f"{number} is a not Prime Number")
            break
    else:
        return True
        print(f"{number} is a Prime Number")
```

### Prime Number Visualisation

<br> Plotting Prime Numbers 

```python
import matplotlib.pyplot as plt
import numpy as np

n = 109
Prime = []

for i in range(2,n+1):
    if isPrime(i):
        Prime.append(i)
        
print(Prime)
print(len(Prime))
print(Prime[len(Prime)-1])

iterated = np.arange(1, (len(Prime) + 1))
y = ((Prime[len(Prime)-1] - Prime[0])/(len(Prime)))*np.arange(1, (len(Prime)+1)) # Trying to do a Linear Regression if Possible
plt.figure(figsize = (15,10))
plt.xlabel("Numbers")
plt.ylabel("Prime Number")
plt.plot(iterated,y)
plt.scatter(iterated, Prime)
```
